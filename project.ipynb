## MATH QUIZ

```python
import random
import time

operators = ["+", "-", "*", "/"]
min_value = 1
max_value = 10
max_problems = 10


def generate_problem():
    num1 = random.randint(min_value, max_value)
    num2 = random.randint(min_value, max_value)
    operator = random.choice(operators)

    if operator == "+":
        return f"{num1} + {num2}", num1 + num2
    elif operator == "-":
        return f"{num1} - {num2}", num1 - num2
    elif operator == "*":
        return f"{num1} * {num2}", num1 * num2
    elif operator == "/": 
        num1 = num1 * num2  # Ensure the division results in an integer
            # Ensure no division by zero and integer division
        while num2 == 0:
            num2 = random.randint(min_value, max_value)

        return f"{num1} / {num2}", num1 // num2

wrong = 0
input("Welcome to the Math Quiz! Press Enter to start...")  

print("Let's get started! Solve the following problems:")
print("---------------------------------------------")
strt_time = time.time()  # Start the timer

for i in range(max_problems):
    problem, answer = generate_problem()
    while True:
        guess = input(f"Problem {i + 1}: {problem} = ? (Answer: )")
        star = int(guess)  # Convert the guess to an integer

        if star == answer:
            print("Correct!")
            break
        else:
            print("Incorrect. Try again.")
            wrong += 1

end_time = time.time()  # End the timer
elapsed_time = end_time - strt_time

print(f"Quiz completed! You had {wrong} wrong answers.")    
print("Thank you for playing the Math Quiz!")
print(f"Your total time: {elapsed_time:.2f} seconds")
print("Have a great day!")
print("----------------------------------------------")
```

## SLOT MACHINE

```python
MAX_LINE = 3

MAX_BAIT = 1000
MIN_BAIT = 100

def deposit():
    while True:
        amount = input("Enter the amount to deposit: ")

        if amount.isdigit():
            amount = int(amount)

            if MIN_BAIT <= amount <= MAX_BAIT:
                print(f"You have deposited ${amount}.")
                break
            else:
                print(f"Invalid amount. Please enter a number between ${MIN_BAIT} and ${MAX_BAIT}.")
        else:
            print("Invalid input. Please enter a numeric value.")


def get_number_of_lines():
    while True:
        lines = input("Enter the number of lines to bet on (1-" + str(MAX_LINE) + "): ")

        if lines.isdigit():
            lines = int(lines)

            if 1 <= lines <= MAX_LINE:
                return lines
            else:
                print("Invalid number of lines. Please enter a number between 1 and " + str(MAX_LINE) + ".")
        else:
            print("Invalid input. Please enter a numeric value.")


def main():
    balance = deposit()
    lines = get_number_of_lines()
    print(balance, lines)
    print(f"You have bet on {lines} lines with a balance of ${balance}.")
    print("Good luck!")
    print("Have a great day!")
    print("----------------------------------------------")

main()

```
